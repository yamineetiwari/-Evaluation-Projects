- We observed the presence of outliers. However, we have only removed outliers from the numerical columns and independent variables. Therefore, we will proceed to remove outliers from the `total_generation_gwh` column. 

```python
# Calculate the Z-scores for the 'total_generation_gwh' column
from scipy.stats import zscore
out_feature = df['total_generation_gwh']
z = np.abs(zscore(out_feature))

# Identify the indices where the Z-score is greater than the threshold (3 in this case)
outliers = np.where(z > 3)

# Remove the outliers from the DataFrame
df_cleaned = df[z <= 3]

# Display the cleaned DataFrame
df_cleaned.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
#find the best random State
max_accuracy=0
best_state=0
random_state=range(1,200)
for state in random_state:
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=state)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    acc=r2_score(y_test,pred)
    if acc>max_accuracy:
        max_accuracy=acc
        best_state=state
print("Maximum R2_Score is=>",max_accuracy,"On Random State=>",best_state)
```
